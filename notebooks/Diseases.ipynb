```python
from itertools import chain
symptoms = []
for i in df.columns[1:]:
  unique_values = list(df[i].unique())
  for j in unique_values:
    if (j != 0) and (j not in symptoms) and (j != 'nan'):
      symptoms.append(unique_values)
# coverting 2d list to 1 d list and then keeping unique values
symptoms = list(set(chain.from_iterable(symptoms)))
symptoms_data = {}
for i,j in enumerate(symptoms):
  symptoms_data[i]=j
symptoms_data
```

```python
disease = list(df['Disease'].unique())
```

```python
len(disease)
```

```python
df.shape
```

```python
def mapping(row):
    row = row.to_dict()
    disease = row['Disease']
    symp = []
    for a in list(row.values())[1:]:
        if a is not None:
            symp.append(a)
    keys = []
    for b in symp:
        for key, values in symptoms_data.items():
            if values == str(b):
                keys.append(key)
    return {'Disease': disease, 'Keys': keys}

# Assuming 'df' is your DataFrame
result = df.apply(mapping, axis=1, result_type='expand')

# Add the results to 'df' as new columns
df['dis'] = result['Disease']
df['sym_col'] = result['Keys']
```

```python
df.head()
```

```python
symptoms_data
```

```python
def make_df(row):
    row = row.to_dict()
    columnns = row['sym_col']
    mk_data = []
    for key, val in symptoms_data.items():
        if key in columnns:
            mk_data.append(1)
        else:
            mk_data.append(0)
    return mk_data

# Assuming you have df, symptoms_data, and symptoms defined
data = df.apply(make_df, axis=1)

# Create DataFrame from the generated data
cleaned_data = pd.DataFrame(data.tolist(), columns=symptoms)

# Assuming you have a column 'dis' in your original DataFrame df
cleaned_data['Disease'] = df['dis']
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(df['Disease'])
df['encoded_dis']= y_encoded

X = df.drop(columns = ['Disease','encoded_dis'],axis=1)
y = df['encoded_dis']

xtrain,xtest,ytrain,ytest = train_test_split(X,y_encoded,test_size=0.2,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import tensorflow as tf
from sklearn.ensemble import VotingClassifier
from catboost import CatBoostClassifier
```

```python
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
model_params = {
    LogisticRegression: {'C': [0.1, 1, 10], 'solver': ['liblinear', 'lbfgs']},
    KNeighborsClassifier: {'n_neighbors': [3, 5, 7]},
    DecisionTreeClassifier: {'max_depth': [None, 5, 10]},
    RandomForestClassifier: {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]},
    SVC: {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']},
    GaussianNB: {},
    xgb.XGBClassifier: {'max_depth': [3, 5, 7], 'learning_rate': [0.01, 0.1, 0.2]},
    VotingClassifier: {},
    CatBoostClassifier: {'iterations': [100, 200, 300], 'learning_rate': [0.01, 0.1, 0.2]}
}
```

```python
tuned_models = {}

for model, params in model_params.items():
    if model != VotingClassifier:
        grid_search = GridSearchCV(estimator=model(), param_grid=params, cv=10)
        grid_search.fit(xtrain.values, ytrain)
        tuned_models[model.__name__] = grid_search.best_estimator_
    else:
        estimators = [(f'model{i}', model_instance) for i, model_instance in tuned_models.items()]
        voting_clf = VotingClassifier(estimators=estimators, voting='hard')
        tuned_models['VotingClassifier'] = voting_clf
```

```python
data = [{'Model_Name': model_name, 'Best_Parameter': best_params} for model_name, best_params in tuned_models.items()]
model_training = pd.DataFrame(data)
model_training.to_csv('/content/grid_search.csv')
```

```python
model_training[model_training['Model_Name'] == 'VotingClassifier']
```

```python
from sklearn.metrics import accuracy_score

# Assuming you have a test set: xtest, ytest

# Iterate through the tuned models
for model_name, model_instance in tuned_models.items():
    if model_name != 'VotingClassifier':
        y_pred = model_instance.predict(xtest.values)  # Assuming xtest is your test data
        accuracy = accuracy_score(ytest, y_pred)
        print(f'Accuracy for {model_name}: {accuracy}')
    else:
        # Fit the VotingClassifier on the training data before making predictions
        model_instance.fit(xtrain.values, ytrain)
        y_pred = model_instance.predict(xtest.values)
        accuracy = accuracy_score(ytest, y_pred)
        print(f'Accuracy for {model_name}: {accuracy}')
```

```python
# Assuming you have a validation set: xval, yval

# Initialize a variable to keep track of the best model and its accuracy
best_model = None
best_accuracy = 0.0

# Iterate through the tuned models
for model_name, model_instance in tuned_models.items():
    if model_name != 'VotingClassifier':
        y_pred = model_instance.predict(xtest.values)  # Assuming xval is your validation data
        accuracy = accuracy_score(ytest, y_pred)
        print(f'Accuracy for {model_name}: {accuracy}')

        if accuracy > best_accuracy:
            best_model = model_instance
            best_accuracy = accuracy
    else:
        # Fit the VotingClassifier on the training data before making predictions
        model_instance.fit(xtrain.values, ytrain)
        y_pred = model_instance.predict(xtest.values)
        accuracy = accuracy_score(ytest, y_pred)
        print(f'Accuracy for {model_name}: {accuracy}')

        if accuracy > best_accuracy:
            best_model = model_instance
            best_accuracy = accuracy

# Now you have the best model stored in the variable best_model

print('Best Model :',best_model,'Best Acuracy :',best_accuracy)
```
